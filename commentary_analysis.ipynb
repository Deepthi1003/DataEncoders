```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
from pycricbuzz import Cricbuzz 
import json 
import threading
import re
import emoji 
import pandas as pd 
import schedule,time


def comment():
    c = Cricbuzz()
    comm = c.commentary('30549')
    jdata = json.loads(json.dumps(comm, indent=4, sort_keys=True))
    df = pd.DataFrame(jdata)
    for user in df["commentary"]:
        cleartxt1 = re.compile('<![^<]*>').sub('', user["comm"])
        cleartxt3 = cleartxt1.replace("|","")
        if sid.polarity_scores(cleartxt3)['pos'] > sid.polarity_scores(cleartxt3)['neg']:
            print(emoji.emojize(":grinning_face_with_big_eyes:"),emoji.emojize(':thumbs_up:')," ",user["comm"]," ",user["over"])
        elif sid.polarity_scores(cleartxt3)['pos'] < sid.polarity_scores(cleartxt3)['neg']:
            print("\U0001F61E",emoji.emojize(':thumbs_down:')," ",user["comm"]," ",user["over"])
        elif sid.polarity_scores(cleartxt3)['pos'] == sid.polarity_scores(cleartxt3)['neg']:
            print("\U0001F642"," ",user["comm"]," ",user["over"])
        print("\n----------------------------------------------------------------------------------------------------------------")
    

# t=threading.Timer(2.0, commentary,["30549"])
# t.start()
schedule.every(1).minutes.do(comment)

while True:
    schedule.run_pending()
    time.sleep(1)
```
